# PyCityShools Analysis

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
sdc_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
sdc_df
```

```python
#Count values
sdc_df.count()
```

```python
#Types of values
sdc_df.dtypes
```

```python
#Total Students using dataframe
total_students = sdc_df['Student ID'].count()

#Total Schools using dataframe
total_schools = sdc_df['School ID'].nunique()

#Total Budget using school data to simplify
total_budget = school_data['budget'].sum()
total_budget = float(total_budget)

#Average Math using school data to simplify
average_math = (student_data['math_score'].sum()) / total_students

#Average Math using school data to simplify
average_reading = (student_data['reading_score'].sum()) / total_students

# Math pass using school data to simplify, then get the value using only math colunm
Math_pass = student_data[student_data['math_score']>69].count()
math_pass = Math_pass['math_score']

#Math percentage
math_percentage = (math_pass / total_students)*100

# Reading pass using school data to simplify, then get the value using only reading colunm
Reading_pass = student_data[student_data.reading_score>69].count()
reading_pass = Reading_pass['reading_score']

# Reading percentage
reading_percentage = (reading_pass / total_students)*100

# Overall pass using count and adding the column at the end to get the numeric value
Overall_pass = student_data[(student_data.reading_score>69) & (student_data.math_score>69)].count()['math_score']

#Overall percentage
Overall_percentage = (Overall_pass / total_students) * 100
```

```python
#Use the variable obtained above to make a new data frame that summarize the district data. 

district_summary = {"Total Schools" : [total_schools],
                   "Total Students" : [total_students],
                   "Total Budget" : [total_budget],
                   "Average Math Score" : [average_math],
                   "Average Reading Score" : [average_reading],
                   "% Passing Math" : [math_percentage],
                   "% Passing Reading" : [reading_percentage],
                   "% Overall Passing" : [Overall_percentage]}
district_df = pd.DataFrame(district_summary)
district_df
```

```python
#Format the column for best reading.#Mapping each colunm

district_df["Total Students"] = district_df["Total Students"].map("{:,}".format)
district_df["Total Budget"] = district_df["Total Budget"].map("${:,}".format)
district_df["Average Math Score"] = district_df["Average Math Score"].map("{:,.2f}".format)
district_df["Average Reading Score"] = district_df["Average Reading Score"].map("{:,.2f}".format)
district_df["% Passing Math"] = district_df["% Passing Math"].map("{:,.2f}%".format)
district_df["% Passing Reading"] = district_df["% Passing Reading"].map("{:,.2f}%".format)
district_df["% Overall Passing"] = district_df["% Overall Passing"].map("{:,.2f}%".format)
```

```python
# Set variables:

#Budget per School
per_school_budget = sdc_df.groupby(["school_name"]).mean()["budget"]

# Calculate the total student count per school
per_school_counts = sdc_df["school_name"].value_counts()
# Calculate the total school budget and per capita spending
per_school_capita = per_school_budget / per_school_counts

# Calculate the average test scores
per_school_math = sdc_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = sdc_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Percentage students pass per school

# Get students who passed math per school
score_passing_math = sdc_df[sdc_df['math_score']>=70]

# Get students who passed reading per school
score_passing_reading = sdc_df[sdc_df['reading_score']>=70]

#passing both
score_passing_both = sdc_df[(sdc_df['math_score']>=70) & (sdc_df['reading_score']>=70)]

# Percentage math passing per school
per_school_passing_math = score_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

# Percentage reading passing per school
per_school_passing_reading = score_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

# Percentage overall passing per school
per_school_passing_overall = score_passing_both.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

```

```python
# Convert to DataFrame
per_school_summary = pd.DataFrame({"School Type": school_type,
                                   "Total Students": sdc_df["school_name"].value_counts(),
                                   "Total School Budget": sdc_df.groupby(["school_name"]).mean()["budget"],
                                   "Per Student Budget": (sdc_df.groupby(["school_name"]).mean()["budget"]) / sdc_df["school_name"].value_counts(),
                                   "Average Math Score": sdc_df.groupby(["school_name"]).mean()["math_score"],
                                   "Average Reading Score": sdc_df.groupby(["school_name"]).mean()["reading_score"],
                                   "% Passing Math": score_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100,
                                   "% Passing Reading": score_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100,
                                   "% Overall Passing": score_passing_both.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100 })
#per_school_summary_df = pd.DataFrame(per_school_summary)
per_school_summary
```

### Score by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#create bins in which data will be held
bins = [0, 585, 615, 645, 675]
names = ["<$585", "$585-615", "$615-645", "$645-675"]

#categorize the spendings 
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_capita,bins,labels = names)
per_school_summary
```

```python
per_spending_df=per_school_summary.drop(columns=["Total Students", "Total School Budget","School Type", "Per Student Budget"])
per_spending_df
```

```python
#Grouping by "Spending Ranges(Per Student), using mean and adding style format"
spending_ranges_grouped = per_spending_df.groupby('Spending Ranges (Per Student)').mean()

spending_ranges_grouped
```
